```python
# import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tweepy
from datetime import datetime

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key,
                    consumer_secret,
                    access_token,
                    access_token_secret)
```

```python
# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
# target data from news sources: BBC, CBS, CNN, Fox, and New York Times
target_users = ("@BBC", "@CBS", "@CNN",
                "@FoxNews", "@nytimes")

# List to hold results
results_list = []
sentiments = []

# Loop through all target users
for target in target_users:
    
    # Counter
    counter = 1

    # Variable for holding the oldest tweet
    oldest_tweet = None

    # Variables for holding sentiments
    compound_list = []
    positive_list = []
    negative_list = []
    neutral_list = []

    # Loop through for the 5 users
    for x in range(1):

        # Run search through each news outlet twitter username looking for 100 tweets
        public_tweets = api.user_timeline(
            target, result_type="recent",count=100, max_id=oldest_tweet)

        # Loop through all tweets
        for tweet in public_tweets:

            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]

            # Add each value to the appropriate list
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)
                
            # Set the new oldest_tweet value
            oldest_tweet = tweet["id"] - 1
            
            # Add sentiments for each tweet into a list
            sentiments.append({"User": target,
                "Text": tweet["text"],
                "Date": tweet["created_at"], 
                "Compound": compound,
                "Positive": pos,
                "Negative": neu,
                "Neutral": neg,
                "Tweets Ago": counter})
        
            # Add to counter 
            counter += 1

    # Store the Average Sentiments
    sentiment = {
        "User": target,
        "Compound": np.mean(compound_list),
        "Positive": np.mean(positive_list),
        "Neutral": np.mean(negative_list),
        "Negative": np.mean(neutral_list),
        "Tweet Count": len(compound_list)
    }

    # Print the Average Sentiments
    print(sentiment)
    print()
    
    # Append news results to 'results_list'
    results_list.append(sentiment)
```

```python
# Convert sentiments to DataFrame
sentiments_df = pd.DataFrame(sentiments).set_index("User").round(3)


# Organize df for better reading
sentiments_df = sentiments_df[["Text","Date","Compound","Positive","Neutral","Negative","Tweets Ago"]]
sentiments_df.head()
```

```python
# Create scatter plot of sentiment analysis of media tweets
news_users = ("BBC", "CBS", "CNN", "Fox", "New York Times")
colors = ("lightblue","green","red","blue","yellow")

# for loop to plot each user
for target in target_users:

    x_vals = sentiments_df.loc[target]["Tweets Ago"]
    y_vals = sentiments_df.loc[target]["Compound"]
    plt.scatter(x_vals, y_vals, 
            marker="o",color=colors[target_users.index(target)],alpha=0.7,linewidth=2)

# # Incorporate the other graph properties
now = datetime.now()
date = now.strftime("%m/%d/%Y")
plt.title(f"Sentiment Analysis of Media Tweets ({date})")

plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")
plt.xlim(102,-2)
plt.legend(news_users, title="Media Sources", bbox_to_anchor=(1, 1))
plt.grid(alpha=0.5)

# Save file to output folder
plt.savefig("Output/sentiment_analysis.png")
plt.show()
```

```python
# df of average overall sentiment analysis of media users
news_df = pd.DataFrame(results_list).round(2)
news_df
```

## Analysis

Three trends that were observed through this analysis was that:

There were an abundant of tweets that returned a sentiment anlaysis score of neutral. We can observe that in the Sentiment Analysis of Media Tweets plot on the horizontal line at 0 that the tweet polarity falls on 0 for a lot of the tweets.

Another trend observed was that BBC, CBS, and Fox News had higher overall compound scores compared to CNN and New York Times. Of course that could change given the small sample size with the last 100 tweets.

Moreover, a trend that stands out most is that CBS's tweet polarity is significantly higher than BBC and Fox News. If they were closer in value, it could easily change in terms of higher tweet polarity, but this shows that CBS's individual compound scores were the most positive of the news sources.
